```python
import pandas as pd
names=["id","title","genre"]
movies_data=pd.read_table("ml-1m/movies.dat",sep="::",engine='python',encoding='latin-1',names=names)
movies_data
```

```python
names=["id","gender", "age","occupation","zip_code"]
users_data=pd.read_table("ml-1m/users.dat",sep="::",engine='python',names=names)
users_data
```

```python
names=["userid","movieid","rating","timestamp"]
ratings_data=pd.read_table("ml-1m/ratings.dat",sep="::",engine='python',names=names)
ratings_data
```

```python
merged_table0=ratings_data.merge(users_data, left_on='userid',right_on='id')
merged_table0
```

```python
all_data=merged_table0.merge(movies_data,left_on='movieid',right_on='id')
all_data
```

```python
age_category=all_data.age.unique()
sorted(age_category)
```

```python
rating_by_age=all_data.pivot_table('rating',index='title',columns=['age'],aggfunc='mean')
rating_by_age
```
